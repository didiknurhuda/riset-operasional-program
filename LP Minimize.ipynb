```python
from pulp import *
problem = LpProblem('Bee_Jelly', LpMinimize)
```

```python
#Definisikan Variabel di pulp

x = LpVariable('Bee', lowBound=0, cat=LpInteger)
y = LpVariable('Jelly', lowBound=0, cat=LpInteger)
```

```python
#Fungsi Objektif
problem += 100*x + 80*y

#konstrain
problem += 2*x + y >= 8
problem += 2*x + 3*y >= 12
problem += x >= 2
problem += y >= 1
```

```python
#Menampilkan permasalahan maksimasi program linear
print(problem)
```

```python
#menyelesaikan permasalahan dan cetak nilainya

problem.solve()
print("Jumlah Bee: ", x.varValue)
print("Jumlah Jelly: ", y.varValue)
print("Biaya Produksi: ", value(problem.objective))
```

```python
prob = LpProblem('x', LpMinimize)
#Fungsi Objektif
prob += -3*x1 + x2 + x3

#konstrain
prob += x1 - 2*x2 + x3 <= 11
prob += -4*x1 + x2 + 2*x3 >= 3
prob += 2*x1 - x3 == -1
```

```python
prob1 = LpProblem("Nilai_Minimum", LpMinimize)

prob1 += 2*x1 - x2 + 4*x3 

prob1 += 5*x1 + x2 -3*x3 >= -7
prob1 += 2*x1 - 2*x2 + x3 <= 8
```
