```python
from pulp import *
```

```python
karyawan = ['Karyawan A', 'Karyawan B', 'Karyawan C', 'Karyawan D']
jobs = ['1', '2', '3', '4']
```

```python
cost = {'Karyawan A':{'1': 15, '2': 20, '3':18, '4': 22},
       'Karyawan B':{'1': 14, '2': 16, '3':21, '4': 17},
       'Karyawan C':{'1': 25, '2': 20, '3':23, '4': 22},
       'Karyawan D':{'1': 17, '2': 18, '3':18, '4': 16}}
```

```python
rute = [(i, j) for i in karyawan for j in jobs]
```

```python
prob = LpProblem('Penugasan',LpMinimize)
var = LpVariable.dicts('Tugas', (karyawan,jobs), 0, cat=LpInteger)

#Fungsi Objektif
prob += lpSum(var[i][j]*cost[i][j] for (i,j) in rute)

#Fungsi Kendala
for j in jobs:
    prob += lpSum(var[i][j] for i in karyawan) == 1
    
for i in karyawan:
    prob += lpSum(var[i][j] for j in jobs) == 1
```

```python
prob.solve()
```

```python
print("status:", LpStatus[prob.status])
```

```python
for v in prob.variables():
    if v.varValue > 0:
        print(v.name, '=', v.varValue)
```

```python
print("Minimum cost: ", value(prob.objective))
```
