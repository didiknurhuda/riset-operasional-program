# Jawaban Tugas BIG M

Perhitungan ini menggunakan matriks dan memiliki kekurangan untuk menghitung rasio

```python
from pulp import *
```

```python
problem = LpProblem('Minimum', LpMinimize)
```

```python
x1 = LpVariable('x1', lowBound=0)
x2 = LpVariable('x2', lowBound=0)
x3 = LpVariable('x3', lowBound=0)

problem += 4*x1 + 4*x2 + x3

problem += x1 + x2 + x3 <= 2
problem += 2*x1 + x2 <= 3
problem += 2*x1 + x2 + 3*x3 >= 3
```

```python
problem.solve()
```

```python
for k in problem.variables():
    print(k.name, '=', k.varValue)
```

```python
print('Nilai Minimum: ', value(problem.objective))
```
